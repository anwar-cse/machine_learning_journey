```python
from sklearn.linear_model import LinearRegression
import numpy as np

study_houre=np.array([[1],[2],[3],[4]])
marks=np.array([40,50,60,70])

model=LinearRegression()
model.fit(study_houre,marks)

pre=model.predict([[6]])
print(pre)

print("slop:", model.coef_)
print("intercept:", model.intercept_)

```

Visualization

```python
import matplotlib.pyplot as plt

plt.scatter(study_houre,marks,color="red")
plt.plot(study_houre,model.predict(study_houre))
plt.scatter(7,model.predict([[7]]),color="blue")
```
