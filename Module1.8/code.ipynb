decision tree

```python
from sklearn.tree import DecisionTreeClassifier

x=[[1],[2],[3],[4],[5],[6]]
y=[0,0,0,1,1,1]

model=DecisionTreeClassifier()
model.fit(x,y)
print(model.predict([[4]]))
```

K-Nearest Neighbors (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()

model.fit(x,y)
print(model.predict([[4]]))
```
